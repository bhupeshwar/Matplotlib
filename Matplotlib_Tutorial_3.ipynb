# Matplotlib Tutorial 3 - Plot two histograms 

```python
import matplotlib.pyplot as plt #  importing pyplot module from 
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.show()
```

```python
plt.hist(ml_student_age)

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 )

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar')

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar', orientation = 'horizontal' )

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'horizontal', color = 'm' )

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'horizontal', color = 'm' , label = 'ML Student')

plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = 'm' , label = 'ML Student')

plt.hist(py_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'horizontal', color = 'y' , label = 'Python Student')
plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = 'm' , label = 'ML Student')

plt.hist(py_student_age, bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = 'y' , label = 'Python Student')
plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_student_age, py_student_age], bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = ['m', 'y'] , label = ['ML Student', 'Python Student'])

#plt.hist(py_student_age, bins , rwidth = 0.8 , histtype = 'bar',    orientation = 'vertical', color = 'y' , label = 'Python Student')
plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = ['m', 'y'] , label = ['ML Student', 'Python Student'])

#plt.hist(py_student_age, bins , rwidth = 0.8 , histtype = 'bar',    orientation = 'vertical', color = 'y' , label = 'Python Student')
plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
bins = [15,20,25,30,35,40,45]

plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins , rwidth = 0.8 , histtype = 'bar', 
         orientation = 'vertical', color = ['m', 'y'] , label = ['ML Student', 'Python Student'])

#plt.hist(py_student_age, bins , rwidth = 0.8 , histtype = 'bar',    orientation = 'vertical', color = 'y' , label = 'Python Student')
plt.title('ML Students Age Histograms')
plt.xlabel('Students Age Categaory')
plt.ylabel('No. Students Age')
plt.legend()
plt.show()
```
