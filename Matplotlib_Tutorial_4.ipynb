# Matplotlib Tutorial 4 - Plot Two Bar Chart 

```python
import matplotlib.pyplot as plt #  importing pyplot module from 
import numpy as np
from matplotlib import style
```

```python
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students, width = 0.2)
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge')
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge', color = 'y')
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge', color = 'y', edgecolor = 'm')
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge', color = 'y', edgecolor = 'm', linewidth = 5)
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge', color = 'y', edgecolor = 'm', linewidth = 5
       , alpha = 0.9)
```

```python
plt.bar(classes,class1_students, width = 0.2, align = 'edge', color = 'y', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
```

```python
plt.bar(classes,class1_students, width = 0.6, align = 'edge', color = 'b', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
```

```python
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.bar(classes,class1_students, width = 0.6, align = 'edge', color = 'k', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
```

```python
plt.figure(figsize = (16,9))
plt.barh(classes,class1_students, align = 'edge', color = 'k', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
```

```python
plt.figure(figsize = (16,9))
plt.bar(classes,class1_students, align = 'edge', color = 'k', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
plt.title('Bar Chart of Bhupeshwar Pathania Class' , fontsize = 18)
plt.xlabel('Classes' , fontsize = 15)
plt.ylabel('No. of Students' , fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.bar(classes,class1_students, align = 'edge', color = 'b', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
plt.bar(classes,class2_students, align = 'edge', color = 'g', edgecolor = 'm', linewidth = 5
       , alpha = 0.9, linestyle = '--')
plt.title('Bar Chart of Bhupeshwar Pathania Class' , fontsize = 18)
plt.xlabel('Classes' , fontsize = 15)
plt.ylabel('No. of Students' , fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index,class1_students, width , color = 'b',
       label = 'Class 1 Students')
plt.bar(classes_index + width,class2_students, width,  color = 'g',
       label = 'Class 1 Students')
plt.bar(classes_index + width + width,class3_students, width, color = 'y',
       label = 'Class 1 Students')

plt.xticks(classes_index + width, classes, rotation = 20)
plt.title('Bar Chart of Bhupeshwar Pathania Class' , fontsize = 18)
plt.xlabel('Classes' , fontsize = 15)
plt.ylabel('No. of Students' , fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (16,9))

classes_index = np.arange(len(classes))
width = 0.2

plt.barh(classes_index,class1_students, width , color = 'b',
        label = 'Class 1 Students')
plt.barh(classes_index + width,class2_students, width,  color = 'g',
        label = 'Class 2 Students')
plt.barh(classes_index + width + width,class3_students, width, color = 'y',
        label = 'Class 3 Students')
plt.yticks(classes_index + width, classes, rotation = 20)
plt.title('Bar Chart of Bhupeshwar Pathania Class' , fontsize = 18)
plt.ylabel('Classes' , fontsize = 15)
plt.xlabel('No. of Students' , fontsize = 15)
plt.legend()
plt.show()
```
